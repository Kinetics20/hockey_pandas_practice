```python
team_splits.head()
```

```python
team_splits.drop(columns=['tmID'], inplace=True)
```

```python
team_splits.set_index(['year', 'name'], inplace=True)
```

```python
team_splits = team_splits.loc[:, 'SepW':'AprOL'].copy()
```

```python
team_splits.head()
```

```python
months = team_splits.columns.map(lambda x: x[:3])
matrix = team_splits.columns.map(lambda x: x[3:])
```

```python
months
```

```python
matrix
```

```python
multi_index = pd.MultiIndex.from_arrays([months, matrix])
multi_index
```

```python
team_splits.columns = multi_index
team_splits.head()
```

```python
team_splits =  team_splits.stack(level=0, future_stack=True)
```

```python
team_splits.head()
```

```python
team_splits.index.nlevels
```

```python
team_splits = team_splits.swaplevel(1, 2)
```

```python
team_splits
```

```python
team_splits = team_splits.reorder_levels([2, 0, 1])
```

```python
team_splits.head()
```

```python
team_splits.index.names
```

```python
team_splits.index.set_names('months', level=2, inplace=True)
```

```python
team_splits.head()
```
